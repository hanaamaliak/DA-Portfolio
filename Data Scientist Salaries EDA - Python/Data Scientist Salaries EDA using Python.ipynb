### Data Overview

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
df.head()
```

```python
df.info()
```

### Preprocessing

```python
col = ['experience_level', 'employment_type', 'company_size', 'remote_ratio']

for i in col:
    print(pd.unique(df[i]))
```

```python
#change the name of experience_level column
df['experience_level'] = np.where(df['experience_level'] == 'MI', 'Intermediate', df['experience_level'])
df['experience_level'] = np.where(df['experience_level'] == 'SE', 'Senior', df['experience_level'])
df['experience_level'] = np.where(df['experience_level'] == 'EN', 'Entry Level', df['experience_level'])
df['experience_level'] = np.where(df['experience_level'] == 'EX', 'Executive', df['experience_level'])
```

```python
print(pd.unique(df['experience_level']))
```

```python
#change the name of employment_type column
df['employment_type'] = np.where(df['employment_type'] == 'FT', 'Full Time', df['employment_type'])
df['employment_type'] = np.where(df['employment_type'] == 'CT', 'Contract', df['employment_type'])
df['employment_type'] = np.where(df['employment_type'] == 'PT', 'Part Time', df['employment_type'])
df['employment_type'] = np.where(df['employment_type'] == 'FL', 'Freelance', df['employment_type'])
```

```python
print(pd.unique(df['employment_type']))
```

```python
#change the name of company_size column
df['company_size'] = np.where(df['company_size'] == 'L', 'Large', df['company_size'])
df['company_size'] = np.where(df['company_size'] == 'S', 'Small', df['company_size'])
df['company_size'] = np.where(df['company_size'] == 'M', 'Medium', df['company_size'])
```

```python
print(pd.unique(df['company_size']))
```

```python
#change the name of remote_ratio column
df['remote_ratio'] = np.where(df['remote_ratio'] == 0, 'No Remote', df['remote_ratio'])
df['remote_ratio'] = np.where(df['remote_ratio'] == '50', 'Partial Remote', df['remote_ratio'])
df['remote_ratio'] = np.where(df['remote_ratio'] == '100', 'Full Remote', df['remote_ratio'])
```

```python
print(pd.unique(df['remote_ratio']))
```

```python
#input another csv contain country code and location
code = pd.read_csv('country_code.csv')
code.head()
```

```python
#add company_location_name
df2 = pd.merge(df, code, on ='company_location', how ='left')
df2
```

```python
print(pd.unique(df2['company_location_name']))
```
