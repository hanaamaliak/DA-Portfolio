**Data Description:**

1. `longitude`: A measure of how far west a house is; a higher value is farther west
2. `latitude`: A measure of how far north a house is; a higher value is farther north
3. `housing_median_age`: Median age of a house within a block; a lower number is a newer building
4. `total_rooms`: Total number of rooms within a block
5. `total_bedrooms`: Total number of bedrooms within a block
6. `population`: Total number of people residing within a block
7. `households`: Total number of households, a group of people residing within a home unit, for a block
8. `median_income`: Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. `median_house_value`: Median house value for households within a block (measured in US Dollars)
10. `ocean_proximity`: Location of the house w.r.t ocean/sea

### Data Preprocessing

##### Handling Missing Values

```python
df.isna().sum()
```

From the EDA we've done, we know that `total_bedroms` has a positive skew distribution, so we can fill the missing values using median.

```python
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), 
                            inplace=True)
```

```python
df.isna().sum()
```

##### Handling Duplicate Data

```python
print(df.duplicated().sum())
```

##### Handling Outliers

```python
from scipy import stats

print(f"Total rows before handling outliers: {len(df)}")

filtered_entries = np.array([True] * len(df))
for col in num:
    zscore = abs(stats.zscore(df[col]))
    filtered_entries = (zscore < 3) & filtered_entries
    df2 = df[filtered_entries]

print(f"Total rows after handling outliers: {len(df2)}")
```

```python
#checking outliers
plt.figure(figsize=(16, 7))
for i in range(0, len(num)):
    plt.subplot(3, 3, i+1)
    sns.boxplot(df2[num[i]], color='gray')
    plt.tight_layout()
```

##### Handling Features

```python
# handling multicollinearity, drop independent columns with high correlation
df2.drop(['longitude', 'latitude','total_rooms', 'households','total_bedrooms'], 
        axis=1,inplace=True)
df2.columns
```

```python
df2.info()
```

```python
#feature encoding
df_dummies = pd.get_dummies(df2['ocean_proximity'])
df2 = pd.concat([df_dummies, df2], axis = 1)
df2
```

```python
df2.drop('ocean_proximity', axis=1,inplace=True)
```

```python
df2.columns
```

### Modeling

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import MinMaxScaler
for i in ['housing_median_age', 'population', 'median_income']:
    df2[i] = StandardScaler().fit_transform(df2[i].values.reshape(len(df2), 1))

df2.describe()
```

```python
df2
```

```python
x = df2.drop("median_house_value", axis=1).values
y = df2["median_house_value"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=41,test_size=0.2)
```

##### Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)

pred_linreg = linreg.predict(x_test)
```

```python
linreg_score = linreg.score(x_test,y_test)
print("R2 model {:.2%}".format(linreg_score))
```

```python
linreg_df = pd.DataFrame({"Actual Data": y_test, "Predicted Data": pred_linreg})
linreg_df.head()
```

```python
plt.figure(figsize =(14,6))
linreg_df = linreg_df.reset_index()
linreg_df = linreg_df.drop(["index"], axis=1)
plt.plot(linreg_df[:50])
plt.legend(["Actual", "Predicted"])
plt.title("Linear Regressor Model")
plt.show()
```

```python
#Evaluasi Model
print("MSE: %.2f" % mean_squared_error(y_test, pred_linreg))
print("MAE: %.2f" % mean_absolute_error(y_test, pred_linreg))
print("MAPE: %.2f" % mean_absolute_percentage_error(y_test, pred_linreg))
print("RMSE: %.2f" % mean_squared_error(y_test, pred_linreg, squared=False))

#R-squared
print('R2 score: %.2f' % linreg_score)
```

##### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)

pred_rfr = rfr.predict(x_test)
```

```python
rfr_score = rfr.score(x_test,y_test)
print("R2 model {:.2%}".format(rfr_score))
```

```python
rfr_df = pd.DataFrame({"Actual Data": y_test, "Predicted Data": pred_rfr})
rfr_df.head()
```

```python
plt.figure(figsize =(14,6))
rfr_df = rfr_df.reset_index()
rfr_df = rfr_df.drop(["index"], axis=1)
plt.plot(rfr_df[:50])
plt.legend(["Actual", "Predicted"])
plt.title("Random Forest Model")
plt.show()
```

```python
#Evaluasi Model
print("MSE: %.2f" % mean_squared_error(y_test, pred_rfr))
print("MAE: %.2f" % mean_absolute_error(y_test, pred_rfr))
print("MAPE: %.2f" % mean_absolute_percentage_error(y_test, pred_rfr))
print("RMSE: %.2f" % mean_squared_error(y_test, pred_rfr, squared=False))

#R-squared
print('R2 score: %.2f' % rfr_score)
```

##### XGB Regressor

```python
from xgboost import XGBRegressor
xgr = XGBRegressor()
xgr.fit(x_train, y_train)

pred_xgr = xgr.predict(x_test)
```

```python
xgr_score = xgr.score(x_test,y_test)
print("R2 model {:.2%}".format(xgr_score))
```

### Comparing Models

```python
models = ['Linear Regression', 'Random Forest', 'XGB Regressor', 'LGBM Regressor']
data = [[linreg_score, mean_squared_error(y_test, pred_linreg, squared=False), mean_absolute_percentage_error(y_test, pred_linreg)],
        [rfr_score, mean_squared_error(y_test, pred_rfr, squared=False), mean_absolute_percentage_error(y_test, pred_rfr)],
        [xgr_score, mean_squared_error(y_test, pred_xgr, squared=False), mean_absolute_percentage_error(y_test, pred_xgr)],
       [lgbm_score, mean_squared_error(y_test, pred_lgbm, squared=False), mean_absolute_percentage_error(y_test, pred_lgbm)]]
cols = ['R2 Score','RMSE', 'MAPE']
pd.DataFrame(data = data , columns=cols , index = models).sort_values(ascending=False , by=['R2 Score'])
```

### Model Evaluation Result

* Before starting the modeling process, I've done standardization for some features such as `housing_median_age`, `population`, and `median_income`.
* The dataset is divided into 80% train data and 20% test data.
* I used 4 (four) models to predict housing prices, that is Linear Regression, Random Forest Regression, XGB Regressor, and LGBM Regressor.
* Based on those 4 models, we can see that LGBM Regressor model has the highest R2 Score with 61%. If we see the error from all of the models, LGBM Regressor also has the lowest errors whether in RMSE or MAPE.
* The conclusion is that *LGBM Regressor Model* has the best performance compared to the 3 other models I've tried in this project.
